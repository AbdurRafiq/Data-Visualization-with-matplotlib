# Matplotlib example workflow

We Want 
* Line plot
* Scatter Plot
* Bar plot

```python
#import matplotlib and get it ready for ploting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# prepare data

x = [1,2,3,4]
y = [11,22,33,44]

#setup plot
fig,ax = plt.subplots(figsize=(10,10))

# plot data

ax.plot(x,y) # add data

# Customise plot

ax.set(title="Simple Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

#save fig
fig.savefig("Image/sample.png") 
          
```

# Making figure with Numpy arrays

```python
import numpy as np
```

```python
#Create  Some Data
x = np.linspace(0,10,100)
x[:10]
```

```python
# Plot the data and create line plot

fig,ax = plt.subplots(figsize=(10,10))
ax.plot(x,x**2);
```

```python
#Use the same data \
fig,ax = plt.subplots(figsize=(10,10))
ax.scatter(x,np.exp(x));
```

```python
# Another Scatter Plot 

fig,ax = plt.subplots(figsize=(10,8))
ax.scatter(x,np.sin(x));
```

```python
# Make a plot form dictionary

nut_butter_prices = {"Almod butter" : 10,
                      "Peanut butter": 8,
                      "Cashew butter": 12}

fig,ax = plt.subplots(figsize=(10,7))
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax.set(title = "Dan's Nut Butter Store",
      ylabel = "Price ($)");
```

```python
#
fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
```
